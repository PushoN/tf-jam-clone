```python
import pandas as pd
import sys,os
import numpy as np
import keras

from matplotlib import pyplot as plt
from keras.layers import Input, Dense , Activation
from keras.models import Model , Sequential

%matplotlib inline
```

```python

shots_csv = 'successful_shots.csv'

os.system('copy ..\{0} '.format(shots_csv))
shots = pd.read_csv(shots_csv)
```

```python
#print(shots.columns,shots.dtypes)

# normalized data -- 
x = shots["dist"].values.reshape(-1,1).astype(np.float32)
minX = min(x)
maxX = max(x)
X = (x-minX)/(maxX-minX)

# normalized data -- 
y = shots["force"].values.reshape(-1,1).astype(np.float32)
minY = min(y)
maxY = max(y)
Y = (y-minY)/(maxY-minY)

plt.scatter(X,Y,color='black')
X.shape,Y.shape

```

```python
#------------------- Create Model

model = Sequential()
model.add(Dense(10, name='shot_in' , input_dim=1, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, name='shot_out' , activation='tanh'))
```

```python
#-------------- Compile and Run Model

#model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
#model.fit(X, Y, epochs=50, batch_size=10)

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X, Y, epochs=200, batch_size=10)



```

```python
 # -----------------------------------------------------------------------------\n",
 # Evaluate\n",
 # -----------------------------------------------------------------------------\n",
 
 #scores = model.evaluate(X, Y)\n",
 #print(\"-\" * 100)\n",
 #print(\"\\n%s: %.2f%%\" % (model.metrics_names[0], scores[0]*100))\n",
 
 # -----------------------------------------------------------------------------\n",
 # Compare prediction vs groundtruth\n",
 # -----------------------------------------------------------------------------\n",
 
 y_pred =  model.predict(X)
 plt.scatter(X, y_pred, s = 13, c =  'r')
 plt.scatter(X, Y, s = 1, c = 'b')
 plt.show()
```

```python
#------------ Save and convert model from Keras to Tensorflow so we can load it into TensorflowSharp

model_name = 'shotsmodel.h5'
model.save(model_name)
os.system('keras_to_tensorflow.py -input_model_file {0} -graph_def True'.format(model_name))


```
